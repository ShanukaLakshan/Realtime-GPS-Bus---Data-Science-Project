```python
# code to combine 2 csv files

import pandas as pd
import numpy as np

# read in csv files
df1 = pd.read_csv('trip_data_all copy.csv')
df2 = pd.read_csv('trip_data_09_22.csv')


```

```python
df2
```

```python
# combine the 2 csv files
df = pd.concat([df2, df1])
df
```

```python
# remove trip_id
df = df.drop(columns=['trip_id'])
df
```

```python
# reset trip_id in df
df = df.reset_index(drop=True)
df = df.reset_index()
df = df.rename(columns={'index':'trip_id'})
df['trip_id'] = df['trip_id'] + 1
df
```

```python
# write df to csv 'trip_all'
df.to_csv('trip_all.csv', index=False)
```
