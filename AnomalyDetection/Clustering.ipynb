```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path: performance.ipynb
# Read in the data
trip_data = pd.read_csv('performance_data.csv')
```

```python
trip_data.head()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Feature Engineering - Calculate performance metrics
device_performance = trip_data.groupby('device_id').agg({
    'travel_time': 'mean',
    'dwell_time': 'mean',
    'anomaly_score': 'sum'  # You can include more performance metrics
}).reset_index()

# Data Preparation - Create a dataset for clustering
features = device_performance[['travel_time', 'dwell_time', 'anomaly_score']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Choose a Clustering Algorithm
k = 3  # Number of clusters (you can adjust this)
kmeans = KMeans(n_clusters=k, random_state=42)

# Perform Clustering
device_performance['cluster'] = kmeans.fit_predict(scaled_features)

# assign cluster labels to the original dataset

trip_data = trip_data.merge(device_performance[['device_id', 'cluster']], on='device_id')
 
```
